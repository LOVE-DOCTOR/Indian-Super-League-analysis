```python
grouped_cols_forminutesplayed = df.groupby(["tour_name", "team_name", "short_name", "position_short"]).agg({"actual_minutes_played": ["sum"]})

grouped_cols_forminutesplayed

 
```

```python
grouped_cols_forchancescreated = df.groupby(["tour_name", "team_name", "short_name", "position_short"]).agg({"events.chances_created": ["sum"]})

grouped_cols_forchancescreated

```

```python
grouped_cols_forshots = df.groupby(["tour_name", "team_name", "short_name", "position_short"]).agg({"events.shots": ["sum"]})

grouped_cols_forshots
```

```python
grouped_cols_forshotsontarget = df.groupby(["tour_name", "team_name", "short_name", "position_short"]).agg({"events.shots_on_target": ["sum"]})

grouped_cols_forshotsontarget
```

```python
grouped_cols_forassist = df.groupby(["tour_name", "team_name", "short_name", "position_short"]).agg({"events.assists": ["sum"]})

grouped_cols_forassist
```

```python
grouped_cols_forgoodpasses = df.groupby(["tour_name", "team_name", "short_name", "position_short"]).agg({"touches.good_passes": ["sum"]})

grouped_cols_forgoodpasses
```

```python

df["shot_accuracy"] = df['events.shots_on_target']/df['events.shots'] * 100

df["assist_accuracy"] = df["events.assists"]/df["events.chances_created"] * 100

df["minplay_goalratio"] = df["actual_minutes_played"]/df["events.goals"]
df
```

```python
grouped_cols_shotacc = df.groupby(["tour_name", "team_name", "short_name", "position_short"]).agg({"shot_accuracy": ["sum"]})

grouped_cols_shotacc
```

```python
grouped_cols_assistacc = df.groupby(["tour_name", "team_name", "short_name", "position_short"]).agg({"assist_accuracy": ["sum"]})

grouped_cols_assistacc
```

```python
grouped_cols_ratio_of_minutesplayed_to_goal = df.groupby(["tour_name", "team_name", "short_name", "position_short"]).agg({"minplay_goalratio": ["sum"]})

grouped_cols_ratio_of_minutesplayed_to_goal
```
